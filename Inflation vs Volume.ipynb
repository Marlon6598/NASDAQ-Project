```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import matplotlib.dates as mdates
from scipy.stats import linregress
from api_keys import nas_api_key
from functools import reduce
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
from matplotlib.dates import (YEARLY, DateFormatter,
                              rrulewrapper, RRuleLocator, drange)
from matplotlib.dates import AutoDateLocator, AutoDateFormatter, date2num
from decimal import Decimal
```

```python
url = "https://data.nasdaq.com/api/v3/datasets/RATEINF/CPI_USA.json?"
queryURLjson = f"{url}appid={nas_api_key}"

print(queryURLjson)
```

```python
VolumeDataPath = "data/InflationVsVolume .csv"
InflationvVolume = pd.read_csv(VolumeDataPath)
```

```python
InflationvVolumeDF = pd.DataFrame(InflationvVolume)

InflationvVolumeDF.head()
```

```python
x_values = InflationvVolumeDF["Inflation Value"]
y_values = InflationvVolumeDF["Stock Volume FB"]

# Run linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot scatter plot
plt.scatter(x_values,y_values)

# Plot regression line
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")

# Label plot
plt.title("Inflation Rate Vs. Stock Volume Facebook")
plt.xlabel('Inflation Rate')
plt.ylabel('Stock Volume')

# Print r square value
print(f"R squared: {rvalue**2}")

# Show plot
plt.show()
```
